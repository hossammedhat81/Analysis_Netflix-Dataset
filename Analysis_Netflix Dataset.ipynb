# Q.10. What is the maximum duration of a Movie/Show on Netflix ?


```python
# data.head(2)

data.head(2)
```

```python
# data['Duration'].unique()

data.Duration.unique()
```

```python
# data.Duration.dtypes

data.Duration.dtypes
```

### str.split()


```python
data.head(2)
```

```python
# data[['Minutes','Unit']] = data['Duration'].str.split(' ', expand=True)

data[['Minutes', 'Unit']] = data['Duration'].str.split(' ', expand = True)
```

```python
# data.head(2)

data.head(10)
```

### max()


```python
# Check current data type
print(data['Minutes'].dtype)  # object (string)

# Convert to numeric
data['Minutes'] = pd.to_numeric(data['Minutes'], errors='coerce')

# Check new data type
print(data['Minutes'].dtype)  # int64 or float64

# Now get correct maximum
max_minutes = data[data['Unit'] == 'min']['Minutes'].max()
print(f"Maximum duration: {max_minutes} minutes")  # 143
```

```python
data[data['Unit'] == 'min']['Minutes'].min()
```

```python
data[data['Unit'] == 'min']['Minutes'].mean()
```

# Q.14. What is the most popular genre/type across all content?


```python

print("\nTop 10 most popular genres:")
top_genres = data['Type'].value_counts().head(10)
print(top_genres)

```

# Visualize top genres


```python

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
top_genres.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Popular Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
